## Summary of results


We test the localization ratio on the VGG network

```python
import weightwatcher as ww
import torchvision.models as models
import transformers
from transformers import OpenAIGPTModel,GPT2Model

ww.__version__
```

# VGG model

```python
model = models.vgg11(pretrained=True)
watcher = ww.WeightWatcher(model=model) 
watcher.localization_ratio(savedir = 'Localization_VGG')
```

# GPT

```python
gpt_model = OpenAIGPTModel.from_pretrained('openai-gpt')
gpt_model.eval()
watcher = ww.WeightWatcher(model=gpt_model)
watcher.localization_ratio(savedir = 'Localization_GPT')
```

```python
gpt2_model = GPT2Model.from_pretrained('gpt2')
gpt2_model.eval()
watcher = ww.WeightWatcher(model=gpt2_model)
watcher.localization_ratio(savedir = 'Localization_GPT2')
```
